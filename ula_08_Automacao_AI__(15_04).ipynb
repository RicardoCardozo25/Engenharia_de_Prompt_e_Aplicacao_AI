```python
# Sistema de Aquecimento

# Crie um script Python simples para verificar a temperatura de um sistema. Se for maior que 30, exiba um alerta de superaquecimento.

def verificar_temperatura():
    try:
        temperatura = float(input("Digite a temperatura do sistema (°C): "))

        if temperatura > 30:
            print("⚠️ Alerta: Sistema em superaquecimento!")
        else:
            print("✅ Temperatura dentro do normal.")

    except ValueError:
        print("Erro: valor inválido. Digite um número.")

# Executar
verificar_temperatura()
```

# MISSÃO 01 - AUTOMAÇÃO DE ARQUIVOS


### Gerando 15 documentos aleatórios para teste

Este código criará uma pasta `test_files` e preencherá com 15 arquivos de nomes e extensões aleatórias, contendo conteúdo aleatório.

```python
import os
import random
import string

def generate_random_string(length):
    return ''.join(random.choice(string.ascii_letters + string.digits + ' ') for _ in range(length))

# Define o diretório para os arquivos de teste
test_dir = "Arquivos Bagunçados"
os.makedirs(test_dir, exist_ok=True)

# Extensões de arquivo comuns para usar
extensions = ['txt', 'pdf', 'jpg', 'png', 'doc', 'docx', 'xls', 'xlsx', 'csv', 'mp3', 'zip', 'json', 'xml']

print(f"Criando 15 arquivos aleatórios na pasta '{test_dir}'...")

for i in range(15):
    # Gera um nome de arquivo aleatório
    file_name = generate_random_string(random.randint(5, 15)).replace(' ', '_')
    # Escolhe uma extensão aleatória
    file_extension = random.choice(extensions)
    full_file_name = f"{file_name}.{file_extension}"
    file_path = os.path.join(test_dir, full_file_name)

    # Gera conteúdo aleatório para o arquivo
    file_content = generate_random_string(random.randint(100, 500))

    with open(file_path, 'w') as f:
        f.write(file_content)
    print(f"Criado: {full_file_name}")

print("15 arquivos aleatórios criados com sucesso!")
```

# MISSÃO 02 - CONSULTA SIMPLES A APIs

```python
# DESENVOLVER UM SCRIPT QUE CONSULTE UMA API PUBLICA (VIACEP) RETORNE AS INFORMAÇÕES DE ENDEREÇO FORMATADAS DE MANEIRA LEGÍVEL

import requests

def consultar_cep(cep):
    # Remove caracteres inválidos
    cep = cep.replace("-", "").strip()

    if len(cep) != 8 or not cep.isdigit():
        print("CEP inválido.")
        return

    url = f"https://viacep.com.br/ws/{cep}/json/"

    try:
        resposta = requests.get(url)
        dados = resposta.json()

        if "erro" in dados:
            print("CEP não encontrado.")
            return

        # Formatação do endereço
        endereco = (
            f"📍 Endereço:\n"
            f"Rua: {dados.get('logradouro', 'N/A')}\n"
            f"Bairro: {dados.get('bairro', 'N/A')}\n"
            f"Cidade: {dados.get('localidade', 'N/A')} - {dados.get('uf', 'N/A')}\n"
            f"CEP: {dados.get('cep', 'N/A')}"
        )

        print(endereco)

    except requests.exceptions.RequestException:
        print("Erro ao conectar com a API.")

# Entrada do usuário
cep = input("Digite o CEP: ")
consultar_cep(cep)



# Provacado o erro com caracters especiais e informa CEP inválido. Não houve erro na código.
```

# MISSÃO 03 - SISTEMA DE NOTIFICAÇÕES

```python
import time
import random

# CONFIGURAÇÕES
LIMITE_CRITICO = 40  # porcentagem
INTERVALO = 2  # segundos entre verificações
MAX_ITERATIONS = 10 # Número máximo de iterações para o monitoramento (PARA QUE NÃO FICASSE EXECUTANDO SEM PARAR)

def obter_valor_monitorado():
    """
    Simula um valor de monitoramento (ex: CPU, temperatura, etc.)
    Aqui usei random, mas pode integrar com algo real depois.
    """
    return random.randint(0, 100)

def enviar_alerta_console(valor):
    print(f"\n🚨 ALERTA CRÍTICO 🚨")
    print(f"Valor atingido: {valor}% (acima de {LIMITE_CRITICO}%)\n")

def enviar_email_simulado(valor):
    print("📧 Enviando e-mail...")
    print(f"""
    Assunto: ALERTA CRÍTICO

    Atenção,

    Foi identificado um valor crítico no sistema.

    Valor atual: {valor}%
    Limite definido: {LIMITE_CRITICO}%

    Verifique imediatamente.

    """)
    print("📧 E-mail enviado (simulado)\n")

# Nova função para encapsular a lógica de alerta
def _verificar_e_emitir_alerta(valor_atual, limite_critico, alerta_ativo_anterior):
    """
    Verifica se o valor atual excede o limite crítico e, se necessário, emite um alerta.
    Retorna o novo estado do alerta_ativo.
    """
    if valor_atual > limite_critico:
        if not alerta_ativo_anterior:
            enviar_alerta_console(valor_atual)
            enviar_email_simulado(valor_atual)
            return True  # Alerta ativado
        return True # Mantém o alerta ativo se já estava
    else:
        return False # Reset do alerta se o valor está normal

def monitorar():
    print("Iniciando monitoramento...\n")

    alerta_ativo = False  # evita spam de alertas
    for i in range(MAX_ITERATIONS):
        valor = obter_valor_monitorado()
        print(f"Valor atual: {valor}%")

        # Chama a nova função para lidar com a lógica de alerta
        alerta_ativo = _verificar_e_emitir_alerta(valor, LIMITE_CRITICO, alerta_ativo)

        time.sleep(INTERVALO)
    print(f"\nMonitoramento encerrado após {MAX_ITERATIONS} iterações.")

# EXECUÇÃO
monitorar()
```
