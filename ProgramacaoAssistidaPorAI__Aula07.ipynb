```python
# NUMERO PRIMO

import math

def eh_primo(n):
    if n < 2:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True
```

```python
# Inversão de String (com slicing)

def inverte_texto(texto):
    return texto[::-1]

# Exemplo de uso da função:
minha_string = "Ricardo"
print(f"A string original é: {minha_string}")
print(f"A string invertida é: {inverte_texto(minha_string)}")
```

```python
def soma(a, b):
    return a + b

def subtracao(a, b):
    return a - b

def multiplicacao(a, b):
    return a * b

def divisao(a, b):
    if b == 0:
        return "Erro: divisão por zero!"
    return a / b

def calculadora():
    try:
        a = float(input("Digite o primeiro número: "))
        b = float(input("Digite o segundo número: "))

        operacao = input("Escolha a operação (+, -, *, /): ")

        if operacao == "+":
            print("Resultado:", soma(a, b))
        elif operacao == "-":
            print("Resultado:", subtracao(a, b))
        elif operacao == "*":
            print("Resultado:", multiplicacao(a, b))
        elif operacao == "/":
            print("Resultado:", divisao(a, b))
        else:
            print("Operação inválida!")

    except ValueError:
        print("Erro: entrada inválida. Digite apenas números.")

# Executar
calculadora()
```
