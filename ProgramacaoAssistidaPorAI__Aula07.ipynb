```python
import math

def eh_primo(n):
    if n < 2:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True

# --- Exemplos de teste estruturados ---
test_cases = [
    (2, True),    # 2 é o menor número primo
    (7, True),    # Número primo
    (10, False),  # 10 não é primo (divisível por 2 e 5)
    (17, True),   # Número primo
    (21, False),  # 21 não é primo (divisível por 3 e 7)
    (1, False),   # 1 não é primo por definição matemática
    (0, False),   # 0 não é primo
    (-5, False)   # Números negativos não são primos
]

print("--- Testes Automáticos da função eh_primo ---")
for number, expected_prime in test_cases:
    actual_prime = eh_primo(number)
    status = "CORRETO" if actual_prime == expected_prime else f"INCORRETO (Esperado: {expected_prime})"
    print(f"O número {number} é primo? Resultado: {actual_prime} [{status}]")

# --- Opção de Input do Usuário ---
print("\n--- Teste Interativo da função eh_primo ---")
while True:
    try:
        user_input = input("Digite um número inteiro para verificar se é primo (ou 'sair' para encerrar): ")
        if user_input.lower() == 'sair':
            break

        num = int(user_input)
        resultado = eh_primo(num)
        print(f"O número {num} é primo? {resultado}")
    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro ou 'sair'.")

print("Fim do teste interativo.")
```

```python
# Inversão de String (com slicing)

def inverte_texto(texto):
    return texto[::-1]

# Exemplo de uso da função:
minha_string = "Ricardo"
print(f"A string original é: {minha_string}")
print(f"A string invertida é: {inverte_texto(minha_string)}")
```
